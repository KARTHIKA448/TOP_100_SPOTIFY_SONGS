### EXPLORE THE DATASET

Spotify top songs data set contains information about top 100 songs on Spotify between 2010 and 2019.

```python
df.columns
```

**features :**

    title: The title of the music track.
    artist: The artist or musical group associated with the track.
    top genre: The genre that is considered the primary or most prominent for the track.
    year released: The year when the music track was released.
    added: The date when the track was added or included in the dataset.
    bpm: Beats per minute, a measure of the tempo or speed of the music.
    nrgy: Energy level of the music track.
    dnce: Danceability, a measure of how suitable the track is for dancing.
    dB: Decibels, a unit of measurement for the loudness of the track.
    live: A measure of whether the track was recorded in a live performance.
    val: Valence, a measure of the musical positiveness of the track.
    dur: Duration, the length of the music track in seconds.
    acous: Acousticness, a measure of how acoustic the track is.
    spch: Speechiness, a measure of the presence of spoken words in the track.
    pop: Popularity, a measure of the track's popularity.
    top year: The year when the track was considered to be at its peak or "top."
    artist type: The type of artist, whether it's a solo artist or a duo.

### Most Relevant Artists of the Decade

```python
sns.countplot(x='artist', data=df, order=df['artist'].value_counts()[:20].index)
plt.title('Most Relevant Artists of the Decade')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

- the most popular artists have been: `Taylor Swift, Calvin Harris, Drake, Rihanna and Ariana Grande`

```python
artist_decade = df[['artist', 'top year']]
artist_decade = artist_decade.rename(columns={'top year':'top_year'})
def cont_art(ds):
    years = range(2010, 2020)
    a_d = pd.DataFrame()
    for i in years:
        col = ds[ds.top_year == i]['artist'].value_counts()
        col = col[:3]
        col.to_frame()
        col = col.rename(str(i))
        a_d = a_d.append(col)

    return a_d

artist_decades = cont_art(artist_decade)
```

```python
f2 = plt.figure()
colors = ['tomato','lightseagreen', 'Aqua', 'Firebrick', 'DarkViolet', 'Gainsboro', 'Olive', 'Gold', 'Pink', 'Black', 'Blue', 'Red', 'Goldenrod', 'mistyrose', 'tan', 'azure', 'gray', 'yellow', 'white', 'greenyellow', 'rosybrown', 'royalblue', 'bisque']
artist_decades.plot(kind='bar', ax=f2.gca(), stacked=True ,figsize=(10,9), color=colors, title='Most popular artists of the decade', ylabel='Number of songs in the top', xlabel='Years')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

- at the beginning of the decade it was dominated by Rihanna, in the middle of the decade the most popular artist was Calvin Harris and at the end of the decade new artists like Post Malone and Billie Eilish were starting to gain popularity

###### Songs' Mood (Valence) Over Time

```python
sns.lineplot(x='year released', y='val', data=df)
plt.title('Evolution of Songs\' Mood Over Time')
plt.xlabel('Year Released')
plt.ylabel('Valence')
plt.show()
```

- In the middle of the decade there was a clear decrease in the mood of the songs and it continued until the end of the decade, being the year 2019, the year in which the songs would gain positivity again.

###### Evolution of Energy and Danceability

```python
sns.lineplot(x='year released', y='nrgy', data=df, label='Energy')
sns.lineplot(x='year released', y='dnce', data=df, label='Danceability')
plt.title('Evolution of Energy and Danceability Over Time')
plt.xlabel('Year Released')
plt.ylabel('Value')
plt.legend()
plt.show()
```

- As with the mood variable, the energy of the songs also decreased over the decade.
